```python
# Exercise 1

import thinkstats as t
import Pmf as p
import itertools

# Grouping by midpoint of each bin
# class size, number of classes
data = { 7:  8,
         13: 8,
         17: 14,
         22: 4,
         27: 6,
         32: 12,
         37: 8,
         42: 3,
         47: 2 }

# Figure out the distribution from the student's perspective,
# which is biased due to oversampling of large classes
def mkStudentCsizeList(size, ct):
    return [size] * (size * ct)

studentDataLists = [mkStudentCsizeList(size, ct) for size, ct in data.iteritems()]

studentData = list(itertools.chain.from_iterable(studentDataLists))

# Calc means
deanPmf = p.MakePmfFromDict(data)
deanMean = deanPmf.Mean()

studentPmf = p.MakePmfFromList(studentData)
studentMean = studentPmf.Mean()

def unbiasPmf(pmf):
    newpmf = pmf.Copy()
    
    for x, p in newpmf.Items():
        newpmf.Mult(x, 1.0 / x)
    
    newpmf.Normalize()
    return newpmf

print "Mean according to Dean's data: %f" % deanMean
print "Mean according to students' data: %f" % studentMean
print "Mean according to students' data, unbiased: %f" % unbiasPmf(studentPmf).Mean()
```

```python
import urllib

results = 'http://www.coolrunning.com/results/10/ma/Apr25_27thAn_set1.shtml'

def ConvertPaceToSpeed(pace):
    """Converts pace in MM:SS per mile to MPH."""
    m, s = [int(x) for x in pace.split(':')]
    secs = m*60 + s
    mph  = 1.0 / secs * 60 * 60 
    return mph


def CleanLine(line):
    """Converts a line from coolrunning results to a tuple of values."""
    t = line.split()
    if len(t) < 6:
        return None
    
    place, divtot, div, gun, net, pace = t[0:6]

    if not '/' in divtot:
        return None

    for time in [gun, net, pace]:
        if ':' not in time:
            return None

    return place, divtot, div, gun, net, pace


def ReadResults(url=results):
    """Read results from coolrunning and return a list of tuples."""
    results = []
    conn = urllib.urlopen(url)
    for line in conn.fp:
        t = CleanLine(line)
        if t:
            results.append(t)
    return results


def GetSpeeds(results, column=5):
    """Extract the pace column and return a list of speeds in MPH."""
    speeds = []
    for t in results:
        pace = t[column]
        speed = ConvertPaceToSpeed(pace)
        speeds.append(speed)
    return speeds

def biasPmf(pmf, biasSpeed):
    """
    Takes a PMF of the actual distribution of runners speeds,
    returns a biased PMF of speeds from the viewpoint of someone running
    at biasSpeed speed.
    """
    newpmf = pmf.Copy()
    
    for val, prob in newpmf.Items():
        diff = abs(val - biasSpeed)
        newpmf.Mult(val, diff)
    newpmf.Normalize()
    return newpmf

results = ReadResults()
speeds = GetSpeeds(results)
speedpmf = p.MakePmfFromList(speeds, 'speeds')

speed, prob = zip(*speedpmf.Items())
plt.plot(speed, prob, "o")
plt.xlabel("Speed")
plt.ylabel("Probability")

bias = biasPmf(speedpmf, 7.5)
```

```python
def percentileRank(scores, your_score):
    count = 0
    for score in scores:
        if score <= your_score:
            count += 1
    
    percentile_rank = 100.0 * count / len(scores)
    return percentile_rank

def percentile(scores, percentile_rank):
    scores.sort()
    index = percentile_rank * (len(scores) - 1) / 100
    return scores[index]
    
```

```python
import survey
from math import sqrt 
import thinkstats
import Pmf

def livebirths(table):
  livebirths = 0

  for each in table.records:
    thisoutcome = each.outcome

    if thisoutcome != "N/A" and thisoutcome == 1:
      livebirths += 1

  return livebirths

def livebirthsfirst(table):
  """
  Displays live births in two groups, one of first babies and one of others.
  """

  firstbirths = 0
  otherbirths = 0

  for each in table.records:
    thisoutcome = each.outcome
    thisorder = each.birthord

    if thisoutcome != "N/A" and thisoutcome == 1 and thisorder == 1:
      firstbirths += 1

    elif thisoutcome != "N/A" and thisoutcome == 1 and thisorder!= 1:
      otherbirths += 1

    else:
      # Record was N/A
      pass

  return (firstbirths, otherbirths)

def prob_birth(pmf, binstart, binend):
    fracbirths = 0
    for x in range(binstart, binend + 1):
        fracbirths += pmf.Prob(x)
    return fracbirths

def prob_early(pmf):
    return prob_birth(pmf, 0, 37)

def prob_on_time(pmf):
    return prob_birth(pmf, 38, 40)

def prob_late(pmf):
    return prob_birth(pmf, 41, 52)

def prglengths(table, filterfn):
    """
    Return a list of pregnancy lengths. Supply a filtering function for the birth order. 
    """
    prglengths = []
    for each in table.records:
        if each.outcome == 1 and filterfn(each.birthord) and each.prglength in range(0, 51):
            prglengths.append(each.prglength)
    return prglengths

def main():
    table = survey.Pregnancies()
    table.ReadRecords()
    
    pmf_first_babies = Pmf.MakePmfFromList(prglengths(table, lambda x: x == 1))
    pmf_other_babies = Pmf.MakePmfFromList(prglengths(table, lambda x: x > 1))
    pmf_all_babies = Pmf.MakePmfFromList(prglengths(table, lambda x: True))
    
    pearlyfirst = prob_early(pmf_first_babies)
    pearlyother = prob_early(pmf_other_babies)
    pearlyall = prob_early(pmf_all_babies)

    pontimefirst = prob_on_time(pmf_first_babies)
    pontimeother = prob_on_time(pmf_other_babies)
    pontimeall = prob_on_time(pmf_all_babies)
    
    platefirst = prob_late(pmf_first_babies)
    plateother = prob_late(pmf_other_babies)
    plateall = prob_late(pmf_all_babies)
    
    print "First babies probability of early birth: %f" % pearlyfirst
    print "Other babies probability of early birth: %f" % pearlyother
    print "All babies probability of early birth: %f" % pearlyall
    print
    print "First babies probability of timely birth: %f" % pontimefirst
    print "Other babies probability of timely birth: %f" % pontimeother
    print "All babies probability of timely birth: %f" % pontimeall
    print
    print "First babies probability of late birth: %f" % platefirst
    print "Other babies probability of late birth: %f" % plateother
    print "All babies probability of late birth: %f" % plateall
    print
    print "Relative risk of being born on time vs born late for first babies: %f" % (pontimefirst / platefirst)
    print "Relative risk of being born on time vs born late for other babies: %f" % (pontimeother / plateother)
    print "Relative risk of being born on time vs born late for all babies: %f" % (pontimeall / plateall)

if __name__ == '__main__':
    main()
```
